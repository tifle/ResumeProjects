## Functions

If we want to go beyond simple arithmetic, we'll need to create a `function` that can accomplish a certain task. For example, we a need a function $f(x, a, b) = a * x + b$. This is where defining/creating functions come in handy!

Here's how we define a function:

```python
def my_func(x, a, b):
  output =  a * x + b
  return output

# simplified version of my_func
def simple_my_func(x, a, b):
  return a * x + b
```

Let's call/use our function!

```python
my_func(2, 3, 4)
```

```python
simple_my_func(2, 3, 4)
```

Another example, defining a function for greetings!

## Plotting

For plotting (graphs, images, etc.), we're going to use the `plotnine` external library! This library is used in the Python world/industry for creating various plots.

```python
# import the plotnine library
from plotnine import *
from plotnine.data import mtcars # sample dataset

# plots a scatterplot (each data point represents an individual(car))
(ggplot(mtcars, aes(x = "wt", y = "mpg")) + geom_point())
```
